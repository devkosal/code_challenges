```python
class Solution:
    def hammingWeight(self, n: int) -> int:
        def getbits(n):
            p = 0
            if n < 2**p:
                return 0
            while n >= 2**p:
                p+=1            
            n -= 2**(p-1)
            return 1 + getbits(n)
        return getbits(n)
    
        
```
