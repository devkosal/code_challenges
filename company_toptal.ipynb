```python
def solution(N,K):
    """
    N: winnings
    K: max number of times all-in is possible 
    work backwards from the total winnings, N, to calculate the min num of rounds for reaching N. 
    """
    r = 0 # num rounds 
    while N > 1: 
        # if we can still go all in AND if current N is even, go all in.
        if K > 0 and N%2==0:
            N /= 2
            K -= 1
        # otherwise bet 1
        else:
            N -= 1
        r += 1
    return r
```

```python
def solution(A):
    """
    Uses depth first search to find the number of uniquely colored contiguous contries in a NxM grid. 
    A: NxM grid
    """
    # defining n and m as well as checking for empty inputs
    n = len(A)
    if n == 0:
        return 0
    m = len(A[0])
    if m == 0:
        return 0
    
    def dfs(A, i, j, val):
        # if i or j are outside of the index of A OR if have that index has already been visited, break.
        if i < 0 or j < 0 or i>=n or j>=m or A[i][j] == 0:
            return
        # if the current A value is equal to the value of interest
        if A[i][j] == val:
            A[i][j] = 0 # set point as visited
            # perform dfs recursively on adjacent indices
            dfs(A,i+1,j,val)
            dfs(A,i-1,j,val)
            dfs(A,i,j+1,val)
            dfs(A,i,j-1,val)
        
    
    counts = {}
    # iterate through the grid
    for i in range(n):
        for j in range(m):
            val = A[i][j]
            # perform a depth first search on each value
            dfs(A,i,j,val)
            # update the counts dict
            if val in counts.keys():
                counts[val] += 1
            elif val != 0:
                counts[val] = 1
    return sum(counts.values()) if counts else 0
```
