```
-- Write only the SQL statement that solves the problem and nothing else.

FROM fsia
UNION
SELECT companyName, sharePrice * sharesOutstanding AS marketCapitalization
FROM fsib
```

````python
def minimal_number_of_packages(items, available_large_packages, available_small_packages):
    it = items
    lp = available_large_packages
    sp = available_small_packages
    res = 0
    # takes care of large pckgs
    avail_lp = lp*5
    # if fewer or equal lp's are availble
#     print(avail_lp,it)
    if avail_lp <= it:
        it -= avail_lp
        res += avail_lp/5
    # if more lp's are availble
    else:
        avail_lp = (it//5)*5
        it -= avail_lp
        res += avail_lp/5
        
    # small pckgs
    if sp > it or it == 0:
        return res + it
    return -1
    
# print(minimal_number_of_packages(30, 5, 7),f" Ans: {10}")
# print(minimal_number_of_packages(400, 5, 7),f" Ans: {-1}")
# print(minimal_number_of_packages(19, 5, 7),f" Ans: {7}")
# print(minimal_number_of_packages(31, 5, 7),f" Ans: {11}")
````

````python
def sort_by_price_ascending(json_string):
    inp = eval(json_string)
    res = sorted(inp,key=lambda x: (x["price"],x["name"]))
    return str(res).replace(" ","").replace("'",'"')

print(sort_by_price_ascending('[{"name":"eggs","price":1},{"name":"coffee","price":9.99},{"name":"rice","price":4.04}]'))
````
