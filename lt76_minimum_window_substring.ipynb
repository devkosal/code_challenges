```python
def minWindow(s: str, t: str) -> str:
    n = len(s)
    m = len(t)

    if (not s and not t) or m > n: return ""
    found_ans = False

    if s == t or t in s: return t

    best = s
    l = 0
    r = m

    t_hash = dict.fromkeys(set(t),0)
    for c in t:
        t_hash[c] += 1

    sub_hash = dict.fromkeys(set(s),0)
    for c in s[l:r]:
        sub_hash[c] += 1

    def match():
        for k,v in t_hash.items():
            if sub_hash[k] < v:
                return False
        return True

    while l <= n-m+1:
        sub = s[l:r]
        if match():
            sub_hash[s[l]] -= 1
            l += 1
            if len(sub) < len(best): best = sub
            found_ans = True
            while r-l < m and r < n-1: 
                r+=1
                sub_hash[s[r-1]] += 1
        elif r < n:
            r += 1
            sub_hash[s[r-1]] += 1
        else:
            break
    return best if found_ans else ""
```
